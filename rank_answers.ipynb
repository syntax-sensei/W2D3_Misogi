```python
import pandas as pd

# Load generated answers
df = pd.read_csv("q2_reward/generated_answers_raw.csv")
ranked_rows = []

print("Manual Ranking Interface")

for prompt in df["prompt"].unique():
    # Show 4 candidates per prompt
    group = df[df["prompt"] == prompt].copy().reset_index(drop=True)
    
    print(f"\n--- Prompt: {prompt} ---")
    for i, row in group.iterrows():
        print(f"\nCandidate {i+1}:")
        print(row["answer"][:500])  # truncate for readability
    
    # Ask for rankings
    ranks = input("Enter your ranking for these 4 (comma-separated, e.g., 2,1,3,4): ").split(",")
    ranks = [int(rank.strip()) for rank in ranks]

    group["rank"] = ranks
    ranked_rows.append(group)

# Save final ranked answers
ranked_df = pd.concat(ranked_rows, ignore_index=True)
ranked_df.to_csv("q2_reward/answers.csv", index=False)
print("\nSaved ranked answers to q2_reward/answers.csv")
```
